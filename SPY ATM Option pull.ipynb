```python
import pandas as pd
import numpy as np
import yfinance as yf
payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
#print(first_table.head())
sip=np.array(payload.iloc[:,0])
df=pd.DataFrame()
#for i in range(0,len(sip)):
exp='2022-03-25'
for i in range(0,20):
    try:
        stk=sip[i]
        tick= yf.Ticker(stk)
        ch=tick.option_chain(date=exp).calls
        ch=ch[ch.inTheMoney==False]
        current=(tick.history(period=exp).Close[0])
        #ln=len(ch.iloc[:,0])
        #yld="{:.2%}".format((ch.iloc[0,3])/ch.iloc[0,2])
        strike=ch.iloc[0,2]
        premium=ch.iloc[0,3]
        temp=pd.Series([stk,current,strike,premium,exp])
        df=df.append(temp,ignore_index=True)
    except:
        pass
df.columns=['stock','current','strike','premium','date']
df['yield']=df.premium/df.strike
df['delta']=df.strike-df.current
df['deltaP']=df.delta/df.strike
df['yield_adj']=df['yield']-df['deltaP']
print(df)
df.to_csv("yield.csv")
```
